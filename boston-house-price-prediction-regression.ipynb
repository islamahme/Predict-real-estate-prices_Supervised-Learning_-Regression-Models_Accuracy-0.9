## Public Leaderboard
* For all participants, the same 50% of predictions from the test set are assigned to the public leaderboard. The score you see on the public leaderboard reflects your model‚Äôs accuracy on this portion of the test set.

## Private Leaderboard
* The other 50% of predictions from the test set are assigned to the private leaderboard. The private leaderboard is not visible to participants until the competition has concluded. At the end of a competition, we will reveal the private leaderboard so you can see your score on the other 50% of the test data. The scores on the private leaderboard are used to determine the competition winners. Getting Started competitions are run on a rolling timeline so the private leaderboard is never revealed.

```python
class DataPreprocessor:
    def __init__(self, dataframe):
        self.df = dataframe

    def fillna_with_median(self, column_name):
        median_value = self.df[column_name].median()
        self.df[column_name].fillna(median_value, inplace=True)

    def fillna_with_mode(self, column_name):
        mode_value = self.df[column_name].mode().iloc[0]
        self.df[column_name].fillna(mode_value, inplace=True)
        
preprocessor = DataPreprocessor(train_df)

## The numeric data type

# Replace the empty values in the LoanAmount column with the most median (or mean) value
preprocessor.fillna_with_median("LotFrontage")
# Replace the empty values in the CoapplicantIncome column with the most median (or mean) value
preprocessor.fillna_with_median("MasVnrArea")
# Replace the empty values in the Loan_Amount_Term column with the most median (or mean) value
preprocessor.fillna_with_median("GarageYrBlt")
# Replace the empty values in the Loan_Amount_Term column with the most median (or mean) value
preprocessor.fillna_with_median("SalePrice")



## The Catigrical  data type
 
# Replace the empty values in the Dependents column with the most frequent value
preprocessor.fillna_with_mode("MasVnrType")
# Replace the empty values in the Self_Employed column with the most frequent value
preprocessor.fillna_with_mode("Alley")
# Replace the empty values in the Dependents column with the most frequent value
preprocessor.fillna_with_mode("FireplaceQu")
# Replace the empty values in the Self_Employed column with the most frequent value
preprocessor.fillna_with_mode("Electrical")

# Replace the empty values in the Married column with the most frequent value
preprocessor.fillna_with_mode("BsmtQual")
# Replace the empty values in the Gender column with the most frequent value
preprocessor.fillna_with_mode("BsmtFinType2")
# Replace the empty values in the Married column with the most frequent value
preprocessor.fillna_with_mode("BsmtCond")
# Replace the empty values in the Dependents column with the most frequent value
preprocessor.fillna_with_mode("BsmtFinType1")
# Replace the empty values in the Self_Employed column with the most frequent value
preprocessor.fillna_with_mode("BsmtExposure")

# Replace the empty values in the Gender column with the most frequent value
preprocessor.fillna_with_mode("GarageCond")
# Replace the empty values in the Gender column with the most frequent value
preprocessor.fillna_with_mode("GarageFinish")
# Replace the empty values in the Married column with the most frequent value
preprocessor.fillna_with_mode("GarageType")
# Replace the empty values in the Self_Employed column with the most frequent value
preprocessor.fillna_with_mode("GarageQual")

# Replace the empty values in the Gender column with the most frequent value
preprocessor.fillna_with_mode("Fence")
# Replace the empty values in the Married column with the most frequent value
preprocessor.fillna_with_mode("PoolQC")
# Replace the empty values in the Self_Employed column with the most frequent value
preprocessor.fillna_with_mode("MiscFeature")
      
    
#combined_data.dropna(subset=["SalePrice"])    
# Just Make Sure 
null=train_df.isnull().sum()
print(null[null>0])
print('The Dataset contains', train_df.isna().sum().sum(), 
      'missing values and', train_df.duplicated().sum().sum(), 'duplicated values')
```

```python
import numpy as np

def gradient_descent(X, y, learning_rate=0.01, num_iterations=1000):
    m, n = X.shape  # Number of training examples (m) and features (n)
    theta = np.zeros((n, 1))  # Initialize parameters

    for _ in range(num_iterations):
        for i in range(m):
            xi = X[i, :].reshape((1, n))
            yi = y[i]

            # Compute the gradient for each example
            gradient = xi.T.dot(xi.dot(theta) - yi)

            # Update parameters
            theta = theta - learning_rate * gradient

    return theta

# Example usage:
# X_train is your feature matrix, y_train is your target vector
# Make sure to normalize/standardize features and add bias term if needed
# X_train = ...
# y_train = ...
# X_train = np.c_[np.ones((X_train.shape[0], 1)), X_train]  # Add bias term

# Set hyperparameters
learning_rate = 0.01
num_iterations = 1000

# Run gradient descent
theta = gradient_descent(X_train, y_train, learning_rate, num_iterations)
```
